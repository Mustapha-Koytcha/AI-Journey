# Linear Regression


```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.read_csv("/Users/Mustapha\Documents/Projet/Formation IA/AI-Journey/Linear-Regression/Dataset/housing_cleaned.csv")
```

```python
Y = np.array(df["median_house_value"])
X = np.array(df["median_income"])
n = X.shape
print(n)
```

```python
beta = np.zeros(X.shape)

n = X.shape[0]
X_mean = 1/n * np.sum(X)
Y_mean = 1/n * np.sum(Y)

a = np.sum((X-X_mean)*(Y-Y_mean))/np.sum((X-X_mean)**2)
b = Y_mean - a * X_mean

plt.plot(X, Y, '.')
plt.plot(X, a*X+b)
plt.xlabel("Median income k$")
plt.ylabel("Median house value - $")
plt.title("Linear regression between house values and incomes")
plt.grid()
```
